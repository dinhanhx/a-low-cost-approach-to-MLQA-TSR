```python
%cd "full/path/to/codebase"
import json
from pathlib import Path

from PIL import Image
from tqdm import tqdm

from src.nn.object_detection import ObjectDetection
```

```python
object_detection = ObjectDetection()
image_folder = Path("../data/private_test/private_test_images")
# do this again with training data
```

```python
image_object = {}
for i in tqdm(image_folder.glob("*.jpg")):
    if i.name in image_object:
        continue
    image = Image.open(i).convert("RGB")
    results = object_detection.infer_batch([image], threshold=0.4)
    image_object[i.name] = results[0]

    with open("assets/private_test_image_object.json", "w", encoding="utf-8") as f:
        json.dump(image_object, f, ensure_ascii=False, indent=4)
```
